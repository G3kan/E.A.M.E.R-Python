Daha okunabilir ve kısa kod yazmak için kullanılır



```python
import pandas as pd # Veri işleme kütüphanesi
import numpy as np # Sayısal işlemler için kütüphane
import matplotlib.pyplot as plt # Grafik çizimi için kütüphane
import seaborn as sns # İstatistiksel veri görselleştirme kütüphane
```

NumPy — Sayısal Python
Büyük dizi (array) ve matris işlemleri için

Çok hızlıdır, bilimsel hesaplamalarda kullanılır

```python
#kütüphaneyi np olarak yukarıda tanımladık
#np.array() fonksiyonu ile NumPy dizisi oluşturabiliriz
#Örnek bir NumPy dizisi oluşturalım ve ortalamasını hesaplayalım
#numpy documentasyonuna bakabilirsiniz: https://numpy.org/doc/stable/

a = np.array([1, 2, 3])
print(a.mean())  # Ortalama
```

Pandas — Veri Analizi
Tablolarla (DataFrame) çalışmak için

Excel, CSV, SQL gibi verileri işler


```python
# Pandası pd olarak tanımladık
# Pandas kütüphanesi veri analizi için kullanılır
# pandas documentasyonuna bakabilirsiniz: https://pandas.pydata.org/docs/
# DataFrame Nedir?
# Pandas kütüphanesi ile veri analizi yaparken kullanılan temel veri yapısı
# Satırlar ve sütunlardan oluşur, tıpkı bir Excel tablosu
# DataFrame oluşturmak için pd.DataFrame() fonksiyonunu kullanabiliriz
# Pandas DataFrame örneği
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
print(df)
```

Matplotlib — Grafik Çizimi
Temel grafik ve görselleştirme kütüphanesi

Çizgi grafiği, bar grafiği çizebilirsin

```python
# matpotlib kütüphanesini plt olarak tanımladık
# matplotlib kütüphanesine  bakabilirsiniz: https://matplotlib.org/stable/contents.html
# Grafik çizimi için matplotlib kütüphanesini kullanabiliriz

# Matplotlib ile basit bir grafik çizimi
plt.plot([1, 2, 3], [4, 5, 6]) # X ve Y eksenindeki değerler
plt.show() # Grafiği gösterir 
```

Seaborn — Gelişmiş Görselleştirme
Matplotlib üzerine kurulmuştur

Daha şık, istatistiksel grafikler için



```python
#seaborn kütüphanesi ile daha şık, istatistiksel grafikler çizebiliriz
#seaborn kütüphanesine bakabilirsiniz: https://seaborn.pydata.org/
# seaborn kütüphanesini sns olarak tanımladık

sns.histplot([1, 2, 2, 3, 3, 3]) # Histogram çizimi
plt.title('Örnek Histogram') # Başlık ekleme
plt.xlabel('Değerler') # X ekseni etiketi
plt.ylabel('Frekans') # Y ekseni etiketi
plt.show() # Grafiği gösterir
```

```python
sns.lineplot(x=[1,2,3], y=[4, 5, 6]) # Çubuk grafik

plt.title('Örnek Çubuk Grafik') # Başlık ekleme
plt.xlabel('X Değerleri') # X ekseni etiketi
plt.ylabel('Y Değerleri') # Y ekseni etiketi
plt.show() # Grafiği gösterir
```
